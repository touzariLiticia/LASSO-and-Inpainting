## $ \text{ Comparaison entre  la régression linéaire, la régression ridge et l’algorithme du LASSO.}$

 ### 1 - Régression linéaire dans le contexte de la classification plug-in

Le but de la régression linéaire est de trouver une fonction linéaire $f:Rp→R$ qui permette de prédire l'étiquette $y^i$ à partir du vecteur $x_i$. La fonction objectif à optimiser pour ce modèle est la suivante: 
$$ MSE(f_w,E)=\frac{1}{n}\sum_{i=0}^{n}(y^i-w_0-\sum_{j=1}^{d}w_j x^i_j)^2 $$

Afin d'adapter le modèle à la classification binaire, nous considérerons les étiquettes dans $Y = \{-1, +1\}$ et nous utiliserons la fonction de prédiction $sign(f_w(x))$. 

```python
# On crée un modèle de régression linéaire
clf_lr = LinearRegression()

# On entraîne ce modèle sur notre jeu de donnée
clf_lr.fit(trainx, trainy)

# On récupère l'erreur de norme 2 sur le jeu de données test comme baseline
lr_error = np.mean((np.sign(clf_lr.predict(testx)) - testy) ** 2)

print("L'erreur en utilisant la regression linéaire simple: "+str(round(lr_error,4)))

# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(clf_lr.coef_)
print('poids non nul: '+str(sum(np.where(clf_lr.coef_ != 0, 1, 0))))
```

* L'erreur est importante car les deux chiffres sont très similaires et les variables sont donc corrélées. 
* La régression linéaire souffre de certains inconvénients dans des cas particuliers :
     - lorsque les variables sont corrélées.
     - lorsque le nombre de variables est plus rand que le nombre d'observations.
* Dans ces cas, la solution n'est pas unique et les poids sont très variables et il y a un risque de sur-apprentissage. Et l'un des moyens de résoudre ce problème est la régularisation. 

La régularisation permet de limiter le sur-apprentissage, elle consiste à contrôler simultanément l'erreur du modèle sur le jeu de formation et la complexité du modèle. L'idée n'est donc pas de minimiser seulement l'erreur du modèle, mais une nouvelle fonction objective:
$$ L(f_w,E)=MSE(f_w,E) + \alpha Reg(w)$$
Où $Reg$ est une fonction des poids du modèle.

### 2 - Régression ridge

La régression de ridge utilise la régularisation de Tykhonov qui considère une pénalité par le carré de la norme du vecteur de poids.
$$Reg(w)=\sqrt{\sum_{i=0}^{d} w_i^2}$$
Et donc la fonction objective à optimiser est une forme quadratique
$$ L_2(f_w,E)=MSE(f_w,E) + \alpha \sqrt{\sum_{i=0}^{d} w_i^2}$$
Avec $ \alpha > 0 $ 


* Nous pouvons facilement prouver que la régression ridge admet une solution analytique unique.
* étudions l'évolution de la solution de la régression ridge en fonction des coefficients de régularisation 

```python
# le nombre de alpha qu'on va tester
nb_alphas = 300
# les valeurs des alphas
alphas = np.logspace(-5, 5, nb_alphas)

# On crée le model ridge
ridge = Ridge()
# Cette liste va contenir les poids à chaque fois
coefs = []
# On calculer l'erreur a chaque fois
errors = []
for a in alphas:
    # On utilise le ridge avec le alpha courant 
    ridge.set_params(alpha=a)
    ridge.fit(trainx, trainy)
    coefs.append(ridge.coef_)
    # on empile à chaque fois l'erreur quadratique
    errors.append([lr_error, np.mean((ridge.predict(testx) - testy) ** 2)])
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('poids')
plt.title('Les poids en fonction de la régularisation')
plt.axis('tight')
plt.show()
```

* Nous remarquons :
    * plus l'alpha est grand (le terme de régularisation est plus important) plus les poids sont faibles (nul si l'alpha est assez grand)
    * Plus le alpha est petit (le terme d'erreur est plus grand), plus on se rapproche de la régression linéaire simple  
* Ridge permet de réduire l'amplitude des poids et donc principalement utilisé pour éviter la multicollinéarité et le sur-apprentissage.

```python
ax = plt.gca()
ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.title("L'erreur en fonction du alpha")
plt.show()
```

* Afin de trouver la valeur alpha optimale, nous effectuons une validation croisée avec RidgeCV. 

```python
ridge_clf = RidgeCV(alphas=(0.1,1,10,50,100,200,500,1000))
ridge_clf.fit(trainx, trainy)

print("Alpha optimale :"+str(ridge_clf.alpha_))
ridge_error=np.mean((ridge_clf.predict(testx) - testy) ** 2)
print("l'erreur en utilisant ridge: "+str(round(ridge_error,3)))
# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(ridge_clf.coef_)
print('poids non nul: '+str(sum(np.where(ridge_clf.coef_ != 0, 1, 0))))
```

Nous remarquons que la majorité des poids ne sont pas nuls, donc Ridge abaisse les poids mais ile ne les annule pas.
<br>
<b>le résultat est déjà meilleur que celui de la simple régression linéaire</b>

### 2 - Régression LASSO (Least Absolute Shrinkage Selector Operator)

La régression de LASSO utilise la régularisation de Tykhonov qui considère une pénalité par la norme 1 du vecteur de poids.
$$Reg(w)=\sum_{i=0}^{d} |w|$$
Et donc la fonction objective à optimiser est:
$$ L_2(f_w,E)=MSE(f_w,E) + \alpha \sum_{i=0}^{d} |w|$$

LASSO pousse les choses plus loin, il annule les poids et donc il :
* sélectionne les variables : parce que nous contraignons les coordonnées W à avoir des valeurs basses.
* réduit la dimension : parce qu'il réduit les valeurs des poids au point qu'ils soient nuls et donc, si la variable n'est pas nécessaire pour la prédiction, il l'élimine.

Le problème n'est pas convexe, la solution n'est pas unique et une descente en gradient peut donc être utilisée pour l'optimisation.
<br>
Étudions l'évolution de la solution de la régression LASSO en fonction des coefficients de régularisation 

```python
# le nombre de alpha qu'on va tester
nb_alphas = 300
# les valeurs des alphas
alphas = np.logspace(-5, 5, nb_alphas)

# On crée le model ridge
lasso = Lasso(max_iter=10000)
# Cette liste va contenir les poids à chaque fois
coefs = []
# On calculer l'erreur a chaque fois
errors = []
for a in alphas:
    # On utilise le ridge avec le alpha courant 
    lasso.set_params(alpha=a)
    lasso.fit(trainx, trainy)
    coefs.append(lasso.coef_)
    # on empile à chaque fois l'erreur quadratique
    errors.append([lr_error, np.mean((lasso.predict(testx) - testy) ** 2)])
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('poids')
plt.title('Les poids en fonction de la régularisation')
plt.axis('tight')
plt.show()
```

* Nous remarquons :
    * Plus l'alpha est élevé, plus LASSO élimine les variables (élimine toutes les variables pour un alpha suffisamment important) 
    * Plus l'alpha est petit, plus les variables entre une par une
* Si deux variables sont corrélées. L'une sera sélectionnée par le Lasso, l'autre supprimée. C'est également son avantage par rapport à une régression Ridge qui ne sélectionnera pas de variables.

```python
ax = plt.gca()
ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.title("L'erreur en fonction du alpha")
plt.show()
```

* Afin de trouver la valeur alpha optimale, nous effectuons une validation croisée avec LassoCV. 

```python
lasso_clf = LassoCV(alphas=(10,0.1,0.01,0.001))
lasso_clf.fit(trainx, trainy)

print("Alpha optimale :"+str(lasso_clf.alpha_))
lasso_error=np.mean((lasso_clf.predict(testx) - testy) ** 2)
print("l'erreur en utilisant lasso: "+str(round(lasso_error,3)))
# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(lasso_clf.coef_)
print('poids non nul: '+str(sum(np.where(lasso_clf.coef_ != 0, 1, 0))))
```

Comme on peut le voir (et comme prévu), la majorité des poids sont nuls, Lasso ne laisse que les variables qui participent vraiment à la prédiction

<b>Lasso contre Ridge </b>
   * Le lasso a tendance à bien faire s'il y a un petit nombre de paramètres significatifs et que les autres sont proches de zéro ( lorsque seuls quelques prédicteurs influencent réellement la réponse) parce qu'il réduit moins les coefficients restants que Ridge.
   * Ridge est meilleur lorsque les variables sont fortement corrélées,il conserve les variables redondantes, mais les réduit. Le lasso élimine toutes les variables sauf une. 
<br>


Cependant, dans la pratique, nous ne connaissons pas les vraies valeurs des paramètres, de sorte que les deux points précédents sont quelque peu théoriques. Il suffit d'effectuer une validation croisée pour sélectionner le modèle le plus adapté à un cas spécifique.

<b>Régularisation</b><br>
En conclusion, la régularisation fonctionne bien dans les deux cas, et réduit l'erreur totale du modèle en ajoutant le biais qui quantifie la complexité. L'un réduit considérablement l'influence de certaines variables sur le modèle (Ridge) tandis que l'autre (LASSO) peut les supprimer directement (en fixant leur poids à zéro), et donc être parcimonieux.

<b> L’utilité du LASSO </b><br>
Le Lasso est un modèle linéaire qui estime des coefficients sparce. Il est utile dans certains contextes en raison de sa tendance à préférer des solutions avec moins de coefficients non nuls, ce qui réduit effectivement le nombre de caractéristiques dont dépend la solution donnée. Pour cette raison, Lasso et ses variantes sont fondamentales dans le domaine de Compressed Sensing.

* Principal avantage :
    * le lasso offre des modèles interprétables et stables, et une prévision efficace à un coût raisonnable.
    * Sélection de variables
    * Des solutions quand le nombre de variables et superieure au nombre d'observations
