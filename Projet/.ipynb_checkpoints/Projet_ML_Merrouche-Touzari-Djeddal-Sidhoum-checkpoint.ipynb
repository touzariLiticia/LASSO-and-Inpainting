## $ \text{ Comparaison entre  la régression linéaire, la régression ridge et l’algorithme du LASSO.}$

 ### 1 - Régression linéaire dans le contexte de la classification plug-in

Le but de la régression linéaire est de trouver une fonction linéaire $f:Rp→R$ qui permette de prédire l'étiquette $y^i$ à partir du vecteur $x_i$. La fonction objectif à optimiser pour ce modèle est la suivante: 
$$ MSE(f_w,E)=\frac{1}{n}\sum_{i=0}^{n}(y^i-w_0-\sum_{j=1}^{d}w_j x^i_j)^2 $$

Afin d'adapter le modèle à la classification binaire, nous considérerons les étiquettes dans $Y = \{-1, +1\}$ et nous utiliserons la fonction de prédiction $sign(f_w(x))$. 

```python
# On crée un modèle de régression linéaire
clf_lr = LinearRegression()

# On entraîne ce modèle sur notre jeu de donnée
clf_lr.fit(trainx, trainy)

# On récupère l'erreur de norme 2 sur le jeu de données test comme baseline
lr_error = np.mean((np.sign(clf_lr.predict(testx)) - testy) ** 2)

print("L'erreur en utilisant la regression linéaire simple: "+str(round(lr_error,4)))

# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(clf_lr.coef_)
print('poids non nul: '+str(sum(np.where(clf_lr.coef_ != 0, 1, 0))))
```

* L'erreur est importante car les deux chiffres sont très similaires et les variables sont donc corrélées. 
* La régression linéaire souffre de certains inconvénients dans des cas particuliers :
     - lorsque les variables sont corrélées.
     - lorsque le nombre de variables est plus grand que le nombre d'observations.
* Dans ces cas, la solution n'est pas unique, les poids sont très variables et il y a un risque de sur-apprentissage. L'un des moyens pour résoudre ce problème est la régularisation. 

La régularisation permet de limiter le sur-apprentissage, elle consiste à contrôler simultanément l'erreur du modèle sur le jeu d'apprentissage et la complexité du modèle. L'idée n'est donc pas de minimiser seulement l'erreur du modèle, mais une nouvelle fonction objective:
$$ L(f_w,E)=MSE(f_w,E) + \alpha Reg(w)$$
Où $Reg$ est une fonction des poids du modèle.

### 2 - Régression Ridge

La régression de ridge utilise la régularisation de Tykhonov qui considère une pénalité par le carré de la norme du vecteur de poids.
$$Reg(w)=\sqrt{\sum_{i=0}^{d} w_i^2}$$
Et donc la fonction objective à optimiser est une forme quadratique
$$ L_2(f_w,E)=MSE(f_w,E) + \alpha \sqrt{\sum_{i=0}^{d} w_i^2}$$
Avec $ \alpha > 0 $ 


* Nous pouvons facilement prouver que la régression ridge admet une solution analytique unique.
* étudions l'évolution de la solution de la régression ridge en fonction des coefficients de régularisation 

```python
# le nombre de alpha qu'on va tester
nb_alphas = 300
# les valeurs des alphas
alphas = np.logspace(-5, 5, nb_alphas)

# On crée le model ridge
ridge = Ridge()
# Cette liste va contenir les poids à chaque fois
coefs = []
# On calculer l'erreur à chaque fois
errors = []
# On calule la norme à chaque fois
norms_Ridge = []
for a in alphas:
    # On utilise le ridge avec le alpha courant 
    ridge.set_params(alpha=a)
    ridge.fit(trainx, trainy)
    coefs.append(ridge.coef_)
    norms_Ridge.append(np.linalg.norm(ridge.coef_, ord=2))
    # on empile à chaque fois l'erreur quadratique
    errors.append([lr_error, np.mean((ridge.predict(testx) - testy) ** 2)])
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('poids')
plt.title('Les poids en fonction alpha')
plt.axis('tight')
plt.show()
```

```python
ax = plt.gca()
ax.plot(alphas, norms_Ridge)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('la norme 2')
plt.title('la norme régularisation en fonction de alpha')
plt.axis('tight')
plt.show()
```

* Nous remarquons :
    * plus l'alpha est grand (le terme de régularisation est plus important) plus les poids sont faibles (nul si l'alpha est assez grand)
    * Plus le alpha est petit (le terme d'erreur est plus grand), plus on se rapproche de la régression linéaire simple  
* Ridge permet de réduire l'amplitude (lentement en comparant avec LASSO) des poids et donc principalement utilisé pour éviter la multicollinéarité et le sur-apprentissage.

```python
ax = plt.gca()
ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.title("L'erreur (en test) en fonction du alpha")
plt.show()
```

* Afin de trouver la valeur alpha optimale, nous effectuons une validation croisée avec RidgeCV. 

```python
ridge_clf = RidgeCV(alphas=(0.1,1,10,50,100,200,500,1000))
ridge_clf.fit(trainx, trainy)

print("Alpha optimale :"+str(ridge_clf.alpha_))
ridge_error=np.mean((ridge_clf.predict(testx) - testy) ** 2)
print("l'erreur en utilisant ridge: "+str(round(ridge_error,3)))
# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(ridge_clf.coef_)
print('poids non nul: '+str(sum(np.where(ridge_clf.coef_ != 0, 1, 0))))
```

Nous remarquons que la majorité des poids ne sont pas nuls, donc Ridge abaisse les poids mais ile ne les annule pas.
<br>
<b>le résultat est déjà meilleur que celui de la simple régression linéaire</b>

### 2 - Régression LASSO (Least Absolute Shrinkage Selector Operator)

La régression de LASSO utilise la régularisation de Tykhonov qui considère une pénalité par la norme 1 du vecteur de poids.
$$Reg(w)=\sum_{i=0}^{d} |w_i|$$
Et donc la fonction objective à optimiser est:
$$ L_2(f_w,E)=MSE(f_w,E) + \alpha \sum_{i=0}^{d} |w_i|$$

LASSO pousse les choses plus loin, il annule les poids et donc il :
* sélectionne les variables : parce que nous contraignons les coordonnées W à avoir des valeurs basses.
* réduit la dimension : parce qu'il réduit les valeurs des poids au point qu'ils soient nuls et donc, si la variable n'est pas nécessaire pour la prédiction, il l'élimine.

Le problème n'est pas convexe, la solution n'est pas unique et une descente en gradient peut donc être utilisée pour l'optimisation.
<br>
Étudions l'évolution de la solution de la régression LASSO en fonction des coefficients de régularisation 

```python
# le nombre de alpha qu'on va tester
nb_alphas = 300
# les valeurs des alphas
alphas = np.logspace(-5, 5, nb_alphas)

# On crée le model ridge
lasso = Lasso(max_iter=10000)
# Cette liste va contenir les poids à chaque fois
coefs = []
# On calculer l'erreur a chaque fois
errors = []
# On calule la norme à chaque fois
norms_Lasso = []
for a in alphas:
    # On utilise le ridge avec le alpha courant 
    lasso.set_params(alpha=a)
    lasso.fit(trainx, trainy)
    coefs.append(lasso.coef_)
    norms_Lasso.append(np.linalg.norm(lasso.coef_, ord=1))
    # on empile à chaque fois l'erreur quadratique
    errors.append([lr_error, np.mean((lasso.predict(testx) - testy) ** 2)])
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('poids')
plt.title('Les poids en fonction de alpha')
plt.axis('tight')
plt.show()
```

```python
ax = plt.gca()
ax.plot(alphas, norms_Lasso,label="Norme de lasso")
ax.plot(alphas, norms_Ridge,label="Norme de Ridge")

ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('la norme 1')
plt.title('La norme régularisation de  en fonction de alpha')
plt.axis('tight')
plt.legend()
plt.show()
```

* Nous remarquons :
    * Plus l'alpha est élevé, plus LASSO élimine les variables (élimine toutes les variables pour un alpha suffisamment important) 
    * Plus l'alpha est petit, plus les variables entre une par une
    * Pus alpha est éleve, plus la norme de régularisation diminue très rapidement, ce qui explique la suppression des variables 
* Si deux variables sont corrélées. L'une sera sélectionnée par le Lasso, l'autre supprimée. C'est également son avantage par rapport à une régression Ridge qui ne sélectionnera pas de variables, et c'est pourquoi la norme de LASSO baisse plus vite que celle de Ridge.

```python
ax = plt.gca()
ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.title("L'erreur en test en fonction du alpha")
plt.show()
```

* Afin de trouver la valeur alpha optimale, nous effectuons une validation croisée avec LassoCV. 

```python
lasso_clf = LassoCV(alphas=(10,0.1,0.01,0.001))
lasso_clf.fit(trainx, trainy)

print("Alpha optimale :"+str(lasso_clf.alpha_))
lasso_error=np.mean((lasso_clf.predict(testx) - testy) ** 2)
print("l'erreur en utilisant lasso: "+str(round(lasso_error,3)))
# Affichage de la matrice de poids
plt.title("la matrice des poids")
show_usps(lasso_clf.coef_)
print('poids non nul: '+str(sum(np.where(lasso_clf.coef_ != 0, 1, 0))))
```

Comme on peut le voir (et comme prévu), la majorité des poids sont nuls, Lasso ne laisse que les variables qui participent vraiment à la prédiction

### <b>Lasso contre Ridge</b> 
   * Le lasso a tendance à bien faire s'il y a un petit nombre de paramètres significatifs et que les autres sont proches de zéro ( lorsque seuls quelques prédicteurs influencent réellement la réponse) parce qu'il réduit moins les coefficients restants que Ridge.
   * Ridge est meilleur lorsque les variables sont fortement corrélées,il conserve les variables redondantes, mais les réduit. Le lasso élimine toutes les variables sauf une. 
<br>


Cependant, dans la pratique, nous ne connaissons pas les vraies valeurs des paramètres, de sorte que les deux points précédents sont un peu théoriques. Il suffit d'effectuer une validation croisée pour sélectionner le modèle le plus adapté à un cas spécifique.

```python
# Loading data 
trainx,trainy = load_usps("USPS_train.txt")
testx,testy = load_usps("USPS_test.txt")

# les combinaisons des chiffres qu'on va utiliser pour les classifications bianaires
comb = [(3,8),(5,6),(3,9),(1,2),(4,5),(7,2)]
lasso_error_test=[]
lasso_error_train=[]

ridge_error_test=[]
ridge_error_train=[]

lr_error_test=[]
lr_error_train=[]

lasso_coefs_Null=[]
ridge_coefs_Null= []
lr_coefs_Null= []
for (first, second) in comb:
    
    trainx2,trainy2 = get_two_classes(trainx,trainy,first,second)
    testx2,testy2 = get_two_classes(testx,testy,first,second)
    
    lasso_clf = LassoCV(alphas=(10,0.1,0.01,0.001))
    ridge_clf = RidgeCV(alphas=(0.1,1,10,50,100,200,500,1000))
    lr_clf = LinearRegression()
    lasso_clf.fit(trainx2, trainy2)
    ridge_clf.fit(trainx2, trainy2)
    lr_clf.fit(trainx2, trainy2)
    
    
    lasso_error_test.append(round(np.mean((lasso_clf.predict(testx2) - testy2) ** 2),4))
    lasso_error_train.append(round(np.mean((lasso_clf.predict(trainx2) - trainy2) ** 2),4))
    ridge_error_test.append(round(np.mean((ridge_clf.predict(testx2) - testy2) ** 2),4))
    ridge_error_train.append(round(np.mean((ridge_clf.predict(trainx2) - trainy2) ** 2),4))
    lr_error_test.append(round(np.mean((lr_clf.predict(testx2) - testy2) ** 2),4))
    lr_error_train.append(round(np.mean((lr_clf.predict(trainx2) - trainy2) ** 2),4))
    
    
    lasso_coefs_Null.append(sum(np.where(lasso_clf.coef_ == 0, 1, 0)))
    ridge_coefs_Null.append(sum(np.where(ridge_clf.coef_ == 0, 1, 0)))
    lr_coefs_Null.append(sum(np.where(lr_clf.coef_ == 0, 1, 0)))
    
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
col = [(str(x[0])+' vs '+str(x[1])) for x in comb]
the_table=plt.table(cellText=[lasso_error_train,ridge_error_train,lr_error_train,lasso_error_test,ridge_error_test,lr_error_test, lasso_coefs_Null,ridge_coefs_Null,lr_coefs_Null],
                      rowLabels=['LASSO error train','Ridge error train','LR error train','LASSO error test','Ridge error test','LR error test','NB nul coefs LASSO','NB nul coefs Ridge','NB nul coefs LR'],
                      rowColours= plt.cm.BuPu(np.linspace(0, 0.5, 9)),
                      colLabels= col,loc='center')
the_table.auto_set_font_size(False)
the_table.set_fontsize(24)
the_table.scale(4, 4)

# Removing ticks and spines enables you to get the figure only with table
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
for pos in ['right','top','bottom','left']:
    plt.gca().spines[pos].set_visible(False)
```

<b>Régularisation</b><br>
En conclusion, la régularisation fonctionne bien dans les deux cas, et réduit l'erreur totale du modèle en ajoutant le biais qui quantifie la complexité. L'un réduit considérablement l'influence de certaines variables sur le modèle (Ridge) tandis que l'autre (LASSO) peut les supprimer directement (en fixant leur poids à zéro), et donc être parcimonieux.

<b> L’utilité du LASSO </b><br>
Le Lasso est un modèle linéaire qui estime des coefficients sparce. Il est utile dans certains contextes en raison de sa tendance à préférer des solutions avec moins de coefficients non nuls, ce qui réduit effectivement le nombre de caractéristiques dont dépend la solution donnée. Pour cette raison, Lasso et ses variantes sont fondamentales dans le domaine de Compressed Sensing.

* Les  avantages principaux :
    * le lasso offre des modèles interprétables et stables, et une prévision efficace à un coût raisonnable.
    * Sélection de variables
    * Des solutions quand le nombre de variables et superieure au nombre d'observations

```python
def transforme_valeur(a,i,j,l,k):
    """Transforms value i<a<j into value l<b<k"""
    return l + (((a-i)/(j-i)) * (k-l))
```

```python
def read_im(fn):
    """To read an image"""
    # read image in rgb
    img = plt.imread(fn)
    # convert image from rgb to hsv
    img = mcolors.rgb_to_hsv(img)
    # transfrom values from [0,1] to [-1,+1]
    img = transforme_valeur(img, 0, 1, -1, 1)
    return img
```

```python
file_name = "./base/Avions13.png"
img_test = read_im(file_name)
img_test.shape
print(img_test[0,0])
```

```python
def show_im(img):
    """To display an image"""
    # transfrom values from [-1,+1] to [0,+1]
    img = transforme_valeur(img, -1, 1, 0, 1)
    # convert image from hsv to rgb
    img = mcolors.hsv_to_rgb(img)
    plt.imshow(img)
```

```python
show_im(img_test)
```

```python
def get_patch(i,j,h,im):
    """returns the patch centered in (i,j) of height=h of image=im"""
    # we assume that h isn't a multiple of 2
    r = int(h/2)
    return im[i-r:i+r+1 , j-r:j+r+1]
```

```python
patch_test = get_patch(15,13,5,img_test)
patch_test.shape
```

```python
def patch_to_vector(patch):
    """converts patch(3d) to vector(1d)"""
    # simply flatten the patch
    return patch.flatten()
```

```python
vector_test = patch_to_vector(patch_test)
vector_test.shape
```

```python
def vector_to_patch(vector):
    """converts vector(1d) to patch(3d)"""
    # height of the corresponding patch
    h = int(np.sqrt(vector.shape[0] / 3))
    # simply reshape the vector
    return np.reshape(vector,(h,h,3))
```

```python
patch_reconstructed_test = vector_to_patch(vector_test)
print("shape of the reconstructed patch : ",patch_reconstructed_test.shape)
print("equality between the initial patch and the corresponding patch : ",np.array_equal(patch_test,patch_reconstructed_test))
```

```python
def noise(img,prc):
    """randomly remove (replace value by -100) a percentage of pixels in the image"""

    # generate a boolean mask respecting the given percentage
    mask = np.random.choice([1,0],size=(img.shape[0],img.shape[1]), p=[prc,1-prc]).astype(np.bool)

    # matrix with the same shape as img filled with -100
    replacement = np.full(img.shape, -100)

    # delete (relace with -100) values selcted by the mask
    rslt = np.copy(img)
    rslt[mask] = replacement[mask]
    return rslt
```

```python
# let's regenerate the percentage given as a parameter
print("real percentage : ",(np.count_nonzero(noise(patch_test,0.5) == -100) / 3) / (patch_test.shape[0]*patch_test.shape[1]) )
```

```python
def noise_a_part(i,j,l,c,img,prc):
    """randomly remove (replace value by -100) a percentage of pixels in a part of the image"""

    # generate a boolean mask respecting the given percentage
    mask = np.random.choice([1,0],size=(l,c), p=[prc,1-prc]).astype(np.bool)

    # matrix with the same shape as the part of the img filled with -100
    replacement = np.full((l,c,3), -100)

    # delete (relace with -100) values selcted by the mask
    rslt = np.copy(img)
    rslt[i:i+l,j:j+c][mask] = replacement[mask]
    return rslt
```

```python
show_im(noise_a_part(12,14,50,50,img_test,0.5))
```

```python
def delete_rect(img,i,j,height,width):
    """remove (replace value by -100) a a rectangle from the image"""
    # generate the matrix correspondig to the deleted rectangle (matrix of -100 with same shape as rectangle)
    deletion = np.full((height,width,3), -100)
    # delete (replace elements with -100) in the image
    rslt = np.copy(img)
    rslt[i:i+height, j:j+width] = deletion
    return rslt
```

```python
# let's generate an image with deleted pixels for the sake of testing
img_compromised = noise(img_test,0.2)
show_im(img_compromised)
```

```python
def get_deletedPixels(img_compromised):
    """returns the position of the deleted pixels (= -100) in the image"""
    index_of_deleted = np.where(img_compromised == -100)
    return list(set(zip(index_of_deleted[0], index_of_deleted[1])))
```

```python
def get_patches_indexes(img,h,step):
    """returns the list of pixels representing the patches of height equal to h and step = step"""
    pixels = []
    # we assume that h isn't a multiple of 2
    dep = h//2
    # we compute the pixels
    for i in range(dep,img.shape[0],step):
        for j in range(dep,img.shape[1],step):
            if (((i+dep) < img.shape[0]) and ((j+dep) < img.shape[1])):
                pixels.append((i,j))
    return pixels
```

```python
print("nombre de patchs obtenu avec h = 5 et step = 2 : ",len(get_patches_indexes(img_test,5,2)))
```

```python
def get_copmpromisedPatches_indexes(img_compromised, h, step):
    """
        returns the position of the patches containing deleted pixels
        
        returns two items :
            - sane_patches_indexs : a list containig the pixels in the center of sane patches
            - compromised_patches_indexs :  a list containig the pixels in the center of compromised patches
            
    """

    # The list of pixels representing the patches
    patches_indexes = get_patches_indexes(img_compromised,h,step)
    
    # list of compromised patches(represented by the pixel in the center)
    compromised_patches_indexs = []
    
    # list of sane patches (represented by the pixel in the center)
    sane_patches_indexs = [] 

    
    # we check for each patch if it contains missing pixels
    for px in patches_indexes:
        # we get the patch
        patch = get_patch(px[0], px[1], h, img_compromised)
        # we check if it contains missing values
        if (np.count_nonzero(patch == -100) > 0) :
            # in this case, this patch is added to the list of patches containing missing indexes
            compromised_patches_indexs.append(px)
        else:
            sane_patches_indexs.append(px)
            
    return (compromised_patches_indexs, sane_patches_indexs) 
```

```python
def get_compromised_patches(img_compromised, h, step):
    """returns image patches that contain missing pixels (patches are transformed to vectors)"""
    
    # we recover the pixels in the center of the compromised patches
    compromised,sane = get_copmpromisedPatches_indexes(img_compromised, h, step)
    # we build the patches
    result = [patch_to_vector(get_patch(px[0],px[1],h,img_compromised)) for px in compromised]
    return result
```

```python
compromised_patches_test =  get_compromised_patches(img_compromised, 5, 2)
print(len(compromised_patches_test))
```

```python
def get_dictionary(img_compromised, h, step): 
    """returns the dictionary, i.e. patches with no missing pixels (patches are transformed to vectors)"""
    
    # we recover the pixels in the center of sane patches
    compromised,sane = get_copmpromisedPatches_indexes(img_compromised, h, step)
    # we build the patches
    result = [patch_to_vector(get_patch(px[0],px[1],h,img_compromised)) for px in sane]
    return result
```

##  $\text{Q.2.2.1 Reconstruction d'un patch :}$

```python
def get_weights(patch, dictionnary, alpha = 0.001, iter_max = 1000):
    """
    takes a patch and an input dictionary, and computes (on the dictionary) the vector of weights
    which best approximates the patch (restricted to the expressed pixels) using the LASSO algorithm,
    it also returns the reconstructed patch
    """
    # convert the patch to a vector :
    Vpatch = patch_to_vector(patch)
    
    # get the indexes of expressed pixels
    index_of_expressed = np.where(Vpatch != -100)[0]
    
    # restrict the patch to these pixels
    Vpatch = Vpatch[index_of_expressed]
       
    # data and target must be of same shape
    dictionnary_bar = np.transpose(dictionnary)
    
    # only keep the represented pixels
    dictionnary_bar = dictionnary_bar[index_of_expressed]
    
    # model initialization and fitting
    model = linear_model.Lasso(alpha=alpha, max_iter=iter_max)
    model.fit(dictionnary_bar,Vpatch) 

    
    # reconstruct the patch
    
    reconstructed = patch_to_vector(patch)
    
    # get the indexes of missing pixels
    index_of_deleted = np.where(reconstructed == -100)[0]
    
    # if this patch contains missing pixels that need to be predicted
    if len(index_of_deleted)>0:
        reconstructed[index_of_deleted] = model.predict(np.transpose(dictionnary)[index_of_deleted])
        
    #print(np.count_nonzero(dictionnary == -100))
    #print(Vpatch)
    #print(dictionnary_bar)
    #show_im(vector_to_patch(dictionnary[4]))
    
    # return the weights vector and the reconstructed patch
    return model.coef_,vector_to_patch(reconstructed)
```

## Relation entre la qualité de la reconstruction et le $\alpha$ du LASSO :

```python
################################# Affichage #########################################
plt.figure(3,figsize=(15,15))
i = 0
alphas = [1, 0.5, 0.1, 0.01, 0.001, 0.0001]
norms = []
selcted_features = []
for alpha in alphas:
    parameters, reconstructed = get_weights(noisy_patch_experimentations, incomplete_partly_noisy_dictionnary, alpha= alpha)
    selcted_features.append(np.count_nonzero(parameters != 0))
    norms.append(np.linalg.norm(parameters, ord=1))
    i += 1
    plt.subplot(32*10+i)
    show_im(reconstructed)
    plt.title("alpha du LASSO = "+str(alpha))
```

> Dans ce cas nous faisons la reconstruction du patch sur le dictionnaire incomplet. Il est clair que sur la plage considérée i.e $[1, 0.5, 0.1, 0.01, 0.001, 0.0001]$ plus la valeur de $\alpha$ diminue plus la reconstruction est de meilleure qualité.

## Relation entre le $\alpha$ du LASSO et le nombre de composantes non nulles dans le vecteur de poids $\pmb{w}$ appris :

```python
################################# Affichage #########################################
plt.figure(4,figsize=(10,10))
plt.bar([str(alpha) for alpha in alphas],selcted_features, color='g')
plt.xlabel("alpha du LASSO")
plt.ylabel("nombre de poids non nuls dans W")
plt.title("Nombre de poids dans W différents de 0 en fonction du alpha du LASSO")
```

> Plus le $\alpha$ du LASSO est grand plus le vecteur de poids $\pmb{w}$ contient de zéros et inversement. Quand $\alpha$ est très grand tous les coefficients seront égaux à zéro. Si $\alpha = 0$, il s'agira alors d'une régression linéaire. C'est un processus de sélection où on ne garde que les features les plus importantes "feature selection" . C'est intéressant lorsque nous avons un nombre très important de features, cet algorithme permet de détecter les features qui servent réellement à la prédiction. Ce n'est pas forcément notre cas dans cette application où le dictionnaire peut être petit. Une valeur de $\alpha$ trop grande peut être trop sélective, surtout quand on a un petit dictionnaire ce qui rend très difficile la reconstruction des pixels manquants. Cependant, une valeur de $\alpha$ trop petite peut être trop peu sélective, en effet lorsque nous voulons reconstruire des pixels d'un patch nous voulons le faire à partir des patchs qui lui ressemblent, un processus de sélection est, en effet, nécessaire. Il faut trouver un compromis. Tout ceci explique que sur la plage de $\alpha$ considérée dans les figures précédentes, la reconstruction était de meilleure qualité, on avait besoin d'encore plus d'information pour reconstruire les pixels manquants.

### 2.3.2 Patches ordered from borders to center 

Le parcours des patchs se fait à partir des bordures du rectangle jusqu'au centre de celui-ci.

```python
def Order_IndexPaches(compromised):
    """
    Orders the patches from borders to center
    """
    res=[]
    i=list(set([c[0] for c in compromised]))
    j=list(set([c[1] for c in compromised]))
    tmp=compromised.copy()
    i.sort()
    j.sort()
    while(j!=[] or i!=[]):
        if(i!=[]):
            min_lin=min(i)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][0]==min_lin]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(i[0])
        if(i!=[]):
            max_lin=max(i)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][0]==max_lin]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(i[len(i)-1])
        if(j!=[]):
            min_col=min(j)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][1]==min_col]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(j[0])
        if(j!=[]):
            max_col=max(j)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][1]==max_col]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(j[len(j)-1])
        
    return res 
```

```python
# the incomplete dictionnairy
incomplete_dictionnary = get_dictionary(img_part_deleted, h, step)
print("dictionnaire de l'image bruitée est de taille : ",len(incomplete_dictionnary))

# get index of noisy and sane patches
compromised, sane = get_copmpromisedPatches_indexes(img_part_deleted, h, step)

# order patches
compromised_ordered_borders=Order_IndexPaches(compromised)

#Inpainting
img_restored_2=Inpainting(img_part_deleted,incomplete_dictionnary,compromised_ordered_borders,h)
```

### 2.3.3 Patches ordered from least noisy to most noisy

Le parcours des patchs se par ordre croissant du nombre de pixels supprimés par patch. Ainsi les patchs les moins bruités sont plus prioritaires aux patchs bruités.

```python
def order_patches_byLeastNoisy(compromised,h,img):
    """
    Orders the patches from least noisy to most noisy
    """ 
    nb_deleted_pixel=[]
    for p in compromised:
        patch=get_patch(p[0],p[0],h,img)
        print(p,np.count_nonzero(patch == -100))
        nb_deleted_pixel.append(np.count_nonzero(patch == -100))
    sorted_index=sorted(range(len(nb_deleted_pixel)), key=lambda k: nb_deleted_pixel[k])    
    return [compromised[i] for i in sorted_index]

order_patches_byLeastNoisy(compromised,h,img_part_deleted)
```

### 2.3.4 Patches ordered from borders to center from least noisy to most noisy

Les patchs des bordures du rectangle non encore traité sont récupérés au fur est à mesure et ordonnés par ordre croissant du nombre de pixels supprimés par patch. 

```python
def order_patches_FromBorders_byLeastNoisy(compromised,h,img):
    """
    Orders the patches from borders to center from least noisy to most noisy
    """ 
    ordered_paches=[]
    i=list(set([c[0] for c in compromised]))
    j=list(set([c[1] for c in compromised]))
    tmp=compromised.copy()
    i.sort()
    j.sort()
    while(j!=[] or i!=[]):
        res=[]
        if(i!=[]):
            min_lin=min(i)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][0]==min_lin]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(i[0])
        if(i!=[]):
            max_lin=max(i)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][0]==max_lin]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(i[len(i)-1])
        if(j!=[]):
            min_col=min(j)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][1]==min_col]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(j[0])
        if(j!=[]):
            max_col=max(j)
            l=[tmp[c] for c in range(len(tmp)) if tmp[c][1]==max_col]
            res+=l
            tmp=list(set(tmp).difference(set(l)))
            del(j[len(j)-1])
    
        ordered_paches+=order_patches_byLeastNoisy(res,h,img)
    return ordered_paches
```

```python
def init_confidence_matrix(i,j,width,height,img):
    """
        Returns an initial Confidence Matrix given an image and a deleted part of the image
        (i, j) : beginning of the deleted part
        (width, height ) : of the deleted part
        C(p)=0 if p is a center of a patch that is within the deleted part
        C(p)=1 else
    """
    conf=np.full((img.shape[0],img.shape[1]),1)
    del_part=np.full((width,height),0)
    r1 = int(width/2)
    r1 = int(height/2)
    conf[i:i+width , j:j+height]=del_part
    return conf
```

```python
def update_confidence_matrix(conf_matrix,i,j,width,height):
    conf=conf_matrix.copy()
    new=np.full((width,height),1)
    r1 = int(width/2)
    r1 = int(height/2)
    conf[i-r1:i+r1 , j-r1:j+r1]=new
    return conf
```

```python
def compute_patch_confidence(i,j,h,conf_matrix):
    """
        returns the confidence of a patch 
        (i,j) : center of the patch
        h : side of the patch
    """
    r = int(h/2)
    s=0
    for x in range(i-r,i+r+1):
        for y in range(j-r,j+r+1):
            s+=conf_matrix[x][y]
    return (s/(h*h))
```

```python
def compute_patch_D(i,j,h,edges,img,alpha=255):
    """
    Returns the D factor of a patch 
    (i,j) : center of the patch
    edges : key points in the deleted part of the image
    alpha : is a normalization factor (e.g.,alpha = 255 for a typical grey-level image)
    """
    delta=compute_gradient(i,j,h,img)
    d=1222222
    for p in edges :
        dist= distance.euclidean((i,j), p)
        if (dist<d):
            d=dist
            edge=p
    if(j == edge[1]):
        return 1
    n_p=np.array([abs(i-edge[0]),abs(j-edge[1])])
    
    return (abs(np.dot(delta,n_p.T))/alpha)
    
```

```python
def compute_priority_patch(i,j,h,conf_matrix,edges,img,alpha):
    C=compute_patch_confidence(i,j,h,conf_matrix)
    D=compute_patch_D(i,j,h,edges,img,alpha)
    return C*D
```

```python
def order_Priority_patches(compromised,edges,h,conf_matrix,img,alpha):
    #conf_matrix=init_confidence_matrix(i,j,width,height,img)
    d={}
    for patch in compromised:
        p=compute_priority_patch(patch[0],patch[1],h,conf_matrix,edges,img,alpha)
        d[patch]=p
    ordered_d={k: v for k, v in sorted(d.items(), key=lambda item: item[1],reverse=True)}
    return list(ordered_d.keys())
```
